# Merge imdb dataset

```python
imdb_movies.head()
```

```python
imdb_movies.info()
```

```python
imdb_votes.head()
```

```python
imdb_votes.info()
```

```python
imdb = pd.DataFrame(data={'imdb_title_id':imdb_movies['imdb_title_id'],
                          'title':imdb_movies['title'],
                          'country':imdb_movies['country'],
                          'director':imdb_movies['director'],
                          'cast':imdb_movies['actors'],
                          'original_title':imdb_movies['original_title'],
                          'weighted_average_vote':imdb_votes['weighted_average_vote'],
                          'total_votes':imdb_votes['total_votes']})
                          
imdb.head()
```

```python
imdb.info()
```

```python
merge_index = []

movies = df[df['type'] == 'Movie']
for title in movies['title']:
    if title in imdb['title'].values or title in imdb['original_title'].values:
        imdb_index = imdb[imdb['title'] == title].index
        merge_index.extend(imdb_index)
```

```python
merge_imdb = imdb.iloc[merge_index]

merge_imdb.head()
```

```python
dup_imdb = pd.DataFrame(columns=merge_imdb.columns)

for title in merge_imdb['title']:
    if len(merge_imdb[merge_imdb['title'] == title].index) > 1:
        index = merge_imdb[merge_imdb['title'] == title].index
        highest = merge_imdb.loc[index].sort_values('total_votes', ascending=False).iloc[:1]
        dup_imdb = pd.concat([dup_imdb, highest])
```

```python
dup_imdb = dup_imdb.drop_duplicates(subset=['title'])
merge_imdb = merge_imdb.drop_duplicates(['title'])
```

```python
dup_imdb.head()
```

```python
merge_imdb
```

```python
for title in merge_imdb['title']:
    if title in dup_imdb['title'].values:
        index = merge_imdb[merge_imdb['title'] == title].index[0]
        dup_index = dup_imdb[dup_imdb['title'] == title].index[0]
        for column in merge_imdb.columns:
            merge_imdb.loc[index, column] = dup_imdb.loc[dup_index, column]
```

```python
merge_imdb
```

```python
columns = []

columns.extend(merge_imdb.columns)
columns.extend(df.columns)

new_df = pd.DataFrame(columns=set(columns))
```

```python
for column in merge_imdb.columns:
    new_df[column] = merge_imdb[column]

for title in df['title']:
    if title in new_df['title'].values:
        df_index = df[df['title'] == title].index[0]
        new_index = new_df[new_df['title'] == title].index[0]
        for column in df.columns:
            new_df.loc[new_index, column] = df.loc[df_index, column]
```

### Fill the null value with imdb movies dataset

```python
new_df[new_df['director'].isnull()]['title'].head()
```

```python
new_df[new_df['cast'].isnull()]['title'].head()
```

```python
new_df[new_df['country'].isnull()]['title'].head()
```

```python
null_direct = new_df[new_df['director'].isnull()]
for title in null_direct['title']:
    if title in imdb_movies['title'].values:
        index = new_df[new_df['title'] == title].index[0]
        imdb_index = imdb_movies[imdb_movies['title'] == title].index[0]
        new_df.loc[index, 'director'] = imdb_movies.loc[imdb_index, 'director']
```

```python
null_cast = new_df[new_df['cast'].isnull()]
for title in null_cast['title']:
    if title in imdb_movies['title'].values:
        index = new_df[new_df['title'] == title].index[0]
        imdb_index = imdb_movies[imdb_movies['title'] == title].index[0]
        new_df.loc[index, 'cast'] = imdb_movies.loc[imdb_index, 'actors']
```

```python
null_country = new_df[new_df['country'].isnull()]
for title in null_country['title']:
    if title in imdb_movies['title'].values:
        index = new_df[new_df['title'] == title].index[0]
        imdb_index = imdb_movies[imdb_movies['title'] == title].index[0]
        new_df.loc[index, 'country'] = imdb_movies.loc[imdb_index, 'country']
```

```python
miss_country = ['United Kingdom', 'United States', 'Israel', 'Italy']
null_country = new_df[new_df['country'].isnull()]

for i, country in zip(null_country.index, miss_country):
    new_df.loc[i, 'country'] = country
```
